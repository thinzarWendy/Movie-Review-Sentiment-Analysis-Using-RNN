```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
sent
```

```python
### One Hot Representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word Embedding Representation

from tensorflow.keras.layers import Embedding
#from tensorflow.keras.processing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## feature representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim,input_length=sent_length))
```

```python
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedding_vectors = model.predict(embedded_docs)
embedding_vectors
```

```python
print("Shape of embedding vectors:", embedding_vectors.shape)
# Output shape:
# (number_of_sentences, sentence_length, embedding_dim)
# Example: (7, 8, 10)
# Each word in a sentence now has a 10-dimensional vector.
```
